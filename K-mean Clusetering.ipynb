```python
import pandas as pd

# Loading the dataset
file_path = 'C:/Users/82106/Desktop/데이터분석 프로젝트/개인 프로젝트 1/data/open/preprocessed_sales_df.csv'
sales_data = pd.read_csv(file_path)

# Displaying the first few rows of the dataset to understand its structure
sales_data.head()
```

```python
# Checking for missing values and outliers in the dataset
missing_values = sales_data.isnull().sum().sum()
outliers = sales_data.describe().T[['min', 'max']]

missing_values, outliers.head()
```

```python
# Handling missing values by replacing them with 0
sales_data_filled = sales_data.fillna(0)

# Handling outliers by capping them at the 95th percentile for each column
percentile_95 = sales_data_filled.quantile(0.95)
sales_data_capped = sales_data_filled.clip(upper=percentile_95, axis=1)

# Verifying the changes
sales_data_capped.describe().T[['min', 'max']].head()
```

```python
# Dropping the 'ID' column as it's not needed for clustering
sales_data_for_clustering = sales_data_capped.drop('ID', axis=1)

# Calculating the required features
total_sales = sales_data_for_clustering.sum(axis=1)
average_sales = sales_data_for_clustering.mean(axis=1)
sales_days = (sales_data_for_clustering > 0).sum(axis=1)

# Creating a new DataFrame with these features
clustering_features = pd.DataFrame({
    'Total Sales': total_sales,
    'Average Sales': average_sales,
    'Sales Days': sales_days
})

clustering_features.head()
```

```python
# Reducing the size of the dataset to 25% of its original size for silhouette score calculation
reduced_data_silhouette = clustering_features.sample(frac=0.25, random_state=42)
```

```python
# Function to calculate average intra-cluster distance (compactness)
def average_intra_cluster_distance(data, labels, n_clusters):
    distances = []
    for i in range(n_clusters):
        cluster_data = data[labels == i]
        if not cluster_data.empty:
            cluster_mean = cluster_data.mean(axis=0)
            intra_cluster_distance = (cluster_data - cluster_mean).pow(2).sum(axis=1).pow(0.5).mean()
            distances.append(intra_cluster_distance)
    return distances

# Trying different numbers of clusters and calculating the average intra-cluster distance
intra_cluster_distances = []
for i in range(2, 6):  # Smaller range to reduce computational load
    kmeans = KMeans(n_clusters=i, init='k-means++', random_state=42)
    labels = kmeans.fit_predict(reduced_data_silhouette)
    distances = average_intra_cluster_distance(reduced_data_silhouette, labels, i)
    intra_cluster_distances.append((i, distances))

intra_cluster_distances
```

```python
# Performing K-Means clustering with 4 clusters
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=42)
kmeans.fit(clustering_features)

# Adding the cluster labels to the original dataset
clustering_features['Cluster'] = kmeans.labels_

# Analyzing the centroids of each cluster
centroids = kmeans.cluster_centers_
cluster_centers_df = pd.DataFrame(centroids, columns=clustering_features.columns[:-1])

# Returning the cluster centers and a sample of the dataset with cluster labels
cluster_centers_df, clustering_features.sample(10)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.preprocessing import normalize

# Text preprocessing and TF-IDF transformation
tfidf_vectorizer = TfidfVectorizer(max_features=500, stop_words=None)
tfidf_matrix = tfidf_vectorizer.fit_transform(product_info_data['제품특성'])

# Normalizing the TF-IDF matrix
tfidf_matrix_normalized = normalize(tfidf_matrix)

# Applying K-Means clustering
# Assuming an arbitrary number of clusters for initial analysis
num_clusters = 5
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
kmeans.fit(tfidf_matrix_normalized)

# Adding cluster labels to the original dataframe
product_info_data['Cluster'] = kmeans.labels_

# Displaying a sample of the clustered data
product_info_data.sample(10)
```
